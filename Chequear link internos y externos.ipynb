```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import urllib3
import warnings
import pandas as pd

# Suprimir las advertencias de verificación SSL
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
warnings.simplefilter('ignore', urllib3.exceptions.InsecureRequestWarning)

def get_links(url):
    try:
        response = requests.get(url, verify=False)  # Agregar verify=False para ignorar verificación SSL
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        internal_links = set()
        external_links = set()
        base_url = urljoin(url, '/')
        
        for link in soup.find_all('a', href=True):
            href = link.get('href')
            full_url = urljoin(url, href)
            if full_url.startswith(base_url):
                internal_links.add(full_url)
            else:
                external_links.add(full_url)

        return internal_links, external_links
    except requests.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return set(), set()

def check_links(links):
    results = []
    for link in links:
        try:
            response = requests.head(link, verify=False)  # Agregar verify=False para ignorar verificación SSL
            results.append((link, response.status_code))
        except requests.RequestException as e:
            results.append((link, f"Error: {e}"))
    return results

if __name__ == "__main__":
    url = input("Enter the URL to check: ")
    internal_links, external_links = get_links(url)
    all_links = internal_links | external_links

    print("\nChecking links...\n")
    results = check_links(all_links)

    # Convertir resultados a DataFrame
    df = pd.DataFrame(results, columns=['URL', 'Code Status'])
    print(df)

    # Guardar DataFrame a un archivo CSV
    output_file = 'link_statuses.csv'
    df.to_csv(output_file, index=False)
    print(f"\nResults saved to {output_file}")

    

```
